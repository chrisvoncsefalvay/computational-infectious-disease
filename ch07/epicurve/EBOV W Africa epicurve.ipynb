# Epicurve of the W African EBOV outbreak

```
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

```
ebov_incidence = pd.read_csv("country_timeseries.csv")
```

```
df = ebov_incidence[["Date", "Cases_Guinea", "Cases_Liberia", "Cases_SierraLeone"]].\
    rename(columns={"Cases_Guinea": "Guinea", "Cases_Liberia": "Liberia", "Cases_SierraLeone": "Sierra Leone"}).\
    melt(id_vars="Date", var_name="Country", value_name="Cases")
```

```
df.Date = pd.to_datetime(df.Date)
```

```
fig = plt.figure(figsize=(16, 8))

customPalette = sns.set_palette(sns.color_palette(["#f95d6a", "#2f4b7c", "#a05195"]))

g = sns.lineplot(data=df, x="Date", y="Cases", hue="Country", palette=customPalette)

plt.savefig("ebov_cases.pdf")
```
