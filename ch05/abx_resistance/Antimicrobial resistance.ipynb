# Antimicrobial resistance

```
import numpy as np
from matplotlib import pyplot as plt

COLORS = {"S": "#2f4b7c",
          "E": "#ffa600",
          "I": "#f95d6a",
          "R": "#a05195",
          "D": "#003f5c",
          "C": "#ff7c43",
          "X": "#665191",
          "Y": "#d45087"}
```

```
gamma = 3.0
epsilon_T = 0.4

R0s = np.linspace(0.2, 12, 50)
R0r = np.linspace(0.2, 12, 50)
```

```
def f(r, s):
    beta_r = r * gamma
    beta_s = s * gamma

    return (gamma / epsilon_T) * (1 - beta_s/beta_r) 
```

```
X, Y = np.meshgrid(R0r, R0s)
Z = f(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
